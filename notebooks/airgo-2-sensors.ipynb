# AirGo Testing with Two Sensors

## Introduction
AirGo is designed to add to the layers of protection where masking is difficult to do (e.g. dental work, indoor dining). In the [last testing iteration](https://github.com/Edderic/iaq/blob/main/notebooks/airgo-testing-1-sensor.ipynb), I showed that AirGo with SmartAir's QT3 filter (Airgo QT3) was able to significantly reduce the amount of inhaled aerosols. For example, using a static environment -- placing the AirGo QT3 on the ground and holding a sensor to measure, and measuring with the device turned on and off -- and doing the experiment several times gave an overall fit factor of about 22, which is above the N95 lower limit. However, [Eric Stein](https://twitter.com/toba) expressed concern that the tests done were not realistic for the use cases such as for dental work or indoor dining. He expressed that the device should be tested as closely as possible for the use cases. For example, for the dental work scenario, measurements should be done while tools are moving in the mouth, to get results that generalize to the specific activities. To address this concern, I upgraded my measurement capabilities to be able to do measurements while moving. The more extensive and realistic testing showed more modest results.

## What's New Since Last Time

### Measurement Upgrade
To handle the more realistic testing scenarios, I added another SPS30 sensor from Sensirion for a total of two sensors, to be able to measure simultaneously how dirty the air is in the room and how many times cleaner the air is in the breathing area. I connected the sensors to a Raspberry Pi (a very small, portable computer), hooked up to a USB power bank. I 3D printed an object to hold one of the sensors next to my nose, to measure the concentration in the breathing area. 

<img src="https://breathesafe-airgo.s3.us-east-2.amazonaws.com/images/2-sensors.png" alt="Edderic wearing two sensors" width=500>

When the device receives power and turns on, the Raspberry Pi is programmed to load the `collect_data.py` script to collect the data from both sensors. Every second, the software writes to the `sensor_data_1.csv` and `sensor_data_2.csv` files.

### More models

Since the last post, I also created a different AirGo design, one that uses a custom-cut IKEA Starkvind (99.5% filtration efficiency at 0.3 microns) filter. The point of the design is to be able to accommodate people with large chests. The older design as it currently exists (AirGo QT3) only works for more flat-chested people. I also used a wider filter so that the user can get cleaner air while being able to turn their head somewhat -- the QT3 on filter on the other hand is not very wide, which limits the effectiveness of the device for when the user is turning their head side-to-side.

In this writeup, I'm testing 4 models:

1. AirGo QT3 with visor

2. AirGo QT3 without visor

3. AirGo Starkvind with visor

4. AirGo Starkvind without visor

Note that for the AirGo Starkvind, the visor is still at an early prototype stage (hence me holding it up). Nonetheless, it should be sufficient for testing purposes to help us estimate what a later version could potentially do.

### Testing 

For each of the models mentioned above, I performed [OSHA's 7 exercises](https://www.osha.gov/laws-regs/regulations/standardnumber/1910/1910.134AppA) to test the efficacy of the devices:
    
    (1) Normal breathing. In a normal standing position, without talking, the subject shall breathe normally.
    
    (2) Deep breathing. In a normal standing position, the subject shall breathe slowly and deeply, taking caution so as not to hyperventilate.
    
    (3) Turning head side to side. Standing in place, the subject shall slowly turn his/her head from side to side between the extreme positions on each side. The head shall be held at each extreme momentarily so the subject can inhale at each side.
    
    (4) Moving head up and down. Standing in place, the subject shall slowly move his/her head up and down. The subject shall be instructed to inhale in the up position (i.e., when looking toward the ceiling).
    
    (5) Talking. The subject shall talk out loud slowly and loud enough so as to be heard clearly by the test conductor. The subject can read from a prepared text such as the Rainbow Passage, count backward from 100, or recite a memorized poem or song.
    
    Rainbow Passage
    When the sunlight strikes raindrops in the air, they act like a prism and form a rainbow. The rainbow is a division of white light into many beautiful colors. These take the shape of a long round arch, with its path high above, and its two ends apparently beyond the horizon. There is, according to legend, a boiling pot of gold at one end. People look, but no one ever finds it. When a man looks for something beyond reach, his friends say he is looking for the pot of gold at the end of the rainbow.
    
    (6) Grimace. The test subject shall grimace by smiling or frowning. (This applies only to QNFT testing; it is not performed for QLFT)
    
    (7) Bending over. The test subject shall bend at the waist as if he/she were to touch his/her toes. Jogging in place shall be substituted for this exercise in those test environments such as shroud type QNFT or QLFT units that do not permit bending over at the waist.

In addition, I also added two additional exercises that are more specific to the use cases I outlined: 

    (8) dental work. Use two long objects (pens, pencils, etc.) that are placed in the breathing area to simulate actions of a dental hygienist.

    (9) drinking. Simulate drinking liquids by place a drinking-glass-like object next to the mouth.

The last two are done only for the models without the visor.

# Exploratory Data Analysis

```python
def plot_one_graph(
    graph, 
    breathing_area_data, 
    ambient_data, 
    breathing_area_vars,
    ambient_vars,
    ax=None, 
    i=None, 
    
):
    """
    Parameters:
        graph: dict
            Has a 'window' dict that has a 'start' and an 'end'
        breathing_area_data: pd.DataFrame
            Corresponds to the sensor meant to collect data in the breathing area (by the nose / mouth).
            DataFrame indexed by datetime. Has the <variable> as a column (e.g. 'pm1')
        ambient_data: pd.DataFrame
            Corresponds to the sensor meant to collect data not in the breathing area (by the nose / mouth).
            This is meant to substitute for the counterfactual "What if we didn't have the device turned on? 
            What concentration would the user have breathed in instead?"
            DataFrame indexed by datetime. Has the <variable> as a column (e.g. 'pm1')
        ax: matplotlib.pyplot.axis or matplotlib.axes
            The axis will graph on, or a list of axes.
        i: int
            Integer representing the specific axis will graph on (e.g. ax[i] if ax is a list of matplotlib.pyplot.axis).
        breathing_area_vars: list[str]
            The variables we'll graph from the breathing_area_data dataframe. 
        ambient_vars: list[str]
            The variables we'll graph from the ambient_data dataframe. 
            
    """
    
    if ax is None:
        fig, ax = plt.subplots(1,1)
        
    breathing_area_df = between(str(graph['window']['start']), str(str(graph['window']['end'])), breathing_area_data)
    ambient = between(str(graph['window']['start']), str(str(graph['window']['end'])), ambient_data)
    
    axis = ax
    if i is not None:
        axis = ax[i]

    colors_breathing_area = ['red', 'orange', 'yellow', 'green']
    
    for j, var in enumerate(breathing_area_vars):
        color = colors_breathing_area[j]
        breathing_area_df[[var]].plot(ax=axis, color=color, linestyle='dashed')

    colors_ambient = ['cyan', 'blue', 'purple', 'pink']
    for j, var in enumerate(ambient_vars):
        color = colors_ambient[j]
        ambient[[var]].plot(ax=axis, color=color, linestyle='dashed')

    event_list = []

    for event in graph['events']:
        axis.axvline(event['timedelta'] + graph['window']['start'], color=event['color'])
        event_list.append(event['event'])

    first_to_plot = breathing_area_vars + ambient_vars
    legends = first_to_plot + [e for e in event_list]
    
    axis.legend(legends)
    axis.set_ylabel('Mass concentration (µg/m3)')
    axis.set_title(graph['title'])
```

```python
def plot(metadata, breathing_area_data, ambient_data, breathing_area_vars, ambient_vars, row_size=3):
    fig, ax = plt.subplots(len(metadata),1, figsize=(15, len(metadata) * row_size))

    for i, graph in enumerate(metadata):
        plot_one_graph(
            graph=graph, 
            breathing_area_data=breathing_area_data, 
            ambient_data=ambient_data,
            breathing_area_vars=breathing_area_vars,
            ambient_vars=ambient_vars,
            ax=ax, 
            i=i,
        )
    
    fig.tight_layout()
```

```python
def get_fit_factor_between_two_events(
    event_index, 
    time_of_interest_1, 
    time_of_interest_2, 
    graph_window, 
    events,
    column_1='pm1',
    column_2='pm1'
):
    denominator = time_of_interest_1[
        (time_of_interest_1.index > events[event_index]['timedelta'] + graph_window['start']) \
        & (time_of_interest_1.index < events[event_index + 1]['timedelta'] + graph_window['start'])
    ][column_1].sum()

    numerator = time_of_interest_2[
        (time_of_interest_2.index > events[event_index]['timedelta'] + graph_window['start']) \
        & (time_of_interest_2.index < events[event_index + 1]['timedelta'] + graph_window['start'])
    ][column_2].sum()

    return numerator / denominator
```

```python
collection = []
for graph in metadata:
    sensor_1 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_1)
    sensor_2 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_2)
    events = graph['events']
    for i in range(len(events) - 1):
        event = events[i]
        collection.append({
            'model - visor': graph['title'],
            'event': event['event'],
            'fit_factor': get_fit_factor_between_two_events(
                i, 
                sensor_1, 
                sensor_2, 
                graph['window'], 
                events,
                column_1='pm1 breathing area (adj)',
                column_2='pm1'
                
            )
        })

fit_factors = pd.DataFrame(collection)
```

### AirGo prototypes typically score poorly with exercises that involve a lot of extreme head movement

The first half of the heatmap above is for "dynamic" head movement type of exercises, where there are a lot of head movement (e.g. bending over, side-to-side, up-and-down, walking with turns). Performance in the "static" exercises on the other hand (e.g. deep breathing, grimacing, normal breathing, talking, dental work) are generally higher than in the dynamic regime. One exception is the drinking scenario. The drinking exercise involved placing a glass-like object in between the sensor and the AirGo outlet, and the fit factor dipped to 2, which is in the max of the range for a well-fitting cloth mask. 

### Using a visor helps performance for the AirGo QT3 model

Comparing the AirGo-that-uses-the-SmartAir's-filter-QT3 model (AirGo QT3), performance generally increased. Very notable were performance in the walking with turns exercise and the bending over exercise, a jump from 1.9 to 9.9, and 1 to 7.3 fit factors, comparing without vs. with the visor.

AirGo prototypes (with SmartAir's QT3 filter and IKEA's Starkvind filter) generally perform worse in "dynamic" head movement type of exercise than the "static" head movement type of exercises.

Besides the up-and-down and side-to-side exercises, AirGo QT3 with visor ranged from a fit factor of 6.6 to 24, somewhere around having 2 CR boxes on max (AirKing fan and Grainger Air Handler 20"x20"x2" filters leading to an averaged fit factor of 5 across three sensors, as described in the NIOSH research), and N95-with-no-leaks lower bound performance (fit factor of 20).

### Generally good agreement with the AirGo QT3 model and AirGo Starkvind model without the visor.

Comparing the columns "AirGo QT3 - visor off" and "AirGo Starkvind - visor off", the performance of the rows are very similar to each other. 

### Adding a visor for the AirGo Starkvind sometimes helped, sometimes hurt.

Adding a visor should in theory help because that would help keep the stream of clean air less mixed with the stream of dirty air. 

#### Hurt
Performance generally decreased for the AirGo Starkvind when the visor was used when performance was high in the non-visor setting. Drop from 9.2 to 5.6 for deep breathing, 16 to 5.4 for grimace, and 11 to 7.5 for normal breathing. 

#### Helped

On the other hand, performance went up for talking (4.1 to 8.7). Also, performance increased after introducing the visor when performance was low to begin with (2.4 to 3.7 for bending over, 1.2 to 1.7 for side-to-side, 1.5 to 2.9 for up-and-down, and 1.4 to 3.1 for walking with turns.

#### Possible cause

I think one possible cause for the sometimes-hurting, sometimes-helping phenomenon is that there could be some leaks with the Starkvind filter. I manually cut the Starkvind filter and used a glue gun to glue a border to it to make a custom-fitting filter for the AirGo. I think it's possible that there might be holes that are pushing horizontally along the edges. When the visor is off, the unfiltered air that is pushing horizontally doesn't get to the sensor by the breathing area as much. Using the visor, however, decreases mixing of the clean air with the surrounding air, but the visor may be trapping the unfiltered air that's blowing horizontally, thereby still leading to a less-than-optimal performance.

## Discussion

Note that I'll answer the following questions using the AirGo QT3 model since that seems to be more promising, given its performance in the testing so far.

### Would I use this for eating indoors?

Even though the drinking scenario only gives a fit factor of 2, most of the time when someone is eating, they are usually not drinking anyway. A large amount of time is usually spent talking to other people (4.1 fit factor), eating (maybe a fit factor of 5, if we use the "dental work" as a surrogate, since both involve using some sort of tool / utensil around the mouth area), or being quiet (normal breathing / deep breathing -- 6 to 9.5 fit factors), maybe when watching something, like the TV. If I had to eat indoors with strangers, I would use this tool. It decreases short-range inhalation of aerosols which is good. However, since the fit factors are not in the 100s or 1000s, there is still some non-negligible risk that one could get infected with something as infectious as SARS-CoV-2. Multi layers of protection are still important -- removing the source through testing, decreasing long-range transmission via CR boxes and HEPA filtration, masking with fitted N95s, etc.

### How about for doctor's visits?

Yes! For situations where one doesn't have to keep their mouth accessible, one could use the AirGo QT3 with the visor attached, along with an N95. Together, they can substantially decrease the risk of transmission. Theoretically, the fit factor performance would be multiplicative. For example, let's say you have a leaky N95 (fit factor of 10). Let's say that you are sitting still and breathing normally, which gave a fit factor of around 6.6. Your combined fit factor becomes 66. In other words, the surrounding air could be 66 times dirtier than what you'll be breathing in when you have your leaky N95 and AirGo QT3 with visor, which should strongly decrease your risk. Likewise, if you had a fitted N95 that consistently gives a fit factor of 100, then your combined fit factor becomes 660. Highly protective!

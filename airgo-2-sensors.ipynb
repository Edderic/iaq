```python
def get_fit_factor_between_two_events(event_index, time_of_interest_1, time_of_interest_2, start_datetime, events):
    denominator = time_of_interest_1[
        (time_of_interest_1.index > events[event_index]['timedelta'] + start_datetime) \
        & (time_of_interest_1.index < events[event_index + 1]['timedelta'] + start_datetime)
    ]['pm1'].sum()

    numerator = time_of_interest_2[
        (time_of_interest_2.index > events[event_index]['timedelta'] + start_datetime) \
        & (time_of_interest_2.index < events[event_index + 1]['timedelta'] + start_datetime)
    ]['pm1'].sum()

    return numerator / denominator
```

```python
def get_fit_factor_for_event(event_name, sensor_1_data, sensor_2_data, graph_window, graph_events):
    sensor_1 = between(str(graph_window['start']), str(str(graph_window['end'])), sensor_data_1)
    sensor_2 = between(str(graph_window['start']), str(str(graph_window['end'])), sensor_data_2)
    
    graph_events_df = pd.DataFrame(graph_events)
    event_index = graph_events_df[graph_events_df['event'] == event_name].index[0]
    #import pdb; pdb.set_trace()
    denominator = sensor_1[
        (sensor_1.index > graph_events[event_index]['timedelta'] + graph_window['start']) \
        & (sensor_1.index < graph_events[event_index + 1]['timedelta'] + graph_window['start'])
    ]['pm1'].sum()

    numerator = sensor_2[
        (sensor_2.index > graph_events[event_index]['timedelta'] + graph_window['start']) \
        & (sensor_2.index < graph_events[event_index + 1]['timedelta'] + graph_window['start'])
    ]['pm1'].sum()

    return numerator / denominator
```

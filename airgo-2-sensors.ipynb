```python
fig, ax = plt.subplots(len(metadata),1, figsize=(15, len(metadata) * 3))

for i, graph in enumerate(metadata):
    sensor_1 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_1)
    sensor_2 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_2)
    sensor_1[['pm1']].plot(ax=ax[i], color='r', linestyle='dashed')
    sensor_2[['pm1']].plot(ax=ax[i], linestyle='dashed')

    event_list = []
    for event in graph['events']:
        ax[i].axvline(event['timedelta'] + graph['window']['start'], color=event['color'])
        event_list.append(event['event'])
    
    legends = ['breathing area (pm1)', 'ambient (pm1)'] + [e for e in event_list]
    ax[i].legend(legends)
    ax[i].set_ylabel('Mass concentration (µg/m3)')
    ax[i].set_title(graph['title'])

fig.tight_layout()
```

```python
def get_fit_factor_between_two_events(event_index, time_of_interest_1, time_of_interest_2, graph_window, events):
    denominator = time_of_interest_1[
        (time_of_interest_1.index > events[event_index]['timedelta'] + graph_window['start']) \
        & (time_of_interest_1.index < events[event_index + 1]['timedelta'] + graph_window['start'])
    ]['pm1'].sum()

    numerator = time_of_interest_2[
        (time_of_interest_2.index > events[event_index]['timedelta'] + graph_window['start']) \
        & (time_of_interest_2.index < events[event_index + 1]['timedelta'] + graph_window['start'])
    ]['pm1'].sum()

    return numerator / denominator
```

```python
collection = []
for graph in metadata:
    sensor_1 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_1)
    sensor_2 = between(str(graph['window']['start']), str(str(graph['window']['end'])), sensor_data_2)
    events = graph['events']
    for i in range(len(events) - 1):
        event = events[i]
        collection.append({
            'model - visor': graph['title'],
            'event': event['event'],
            'fit_factor': get_fit_factor_between_two_events(i, sensor_1, sensor_2, graph['window'], events)
        })

fit_factors = pd.DataFrame(collection)
```
